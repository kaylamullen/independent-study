```python
from BaselineBellmanFord import *
import csv
```

```python
graph1 = read_graph("input/grids/full100-edgeweight1.csv")
graphrand = read_graph("input/grids/full100-edgeweightrand.csv")
christinegraph = read_graph('input/christine-inputs/grid100.txt')
```

```python
def testdistance10x(dasher_fname, task_fname, stop_time, graph, results_fname, tasklist_sizes, dasherlist_sizes):
    num_events_per_iter = []
    total_reward_per_iter = []
    data = []
    pct_total_possible_reward_per_iter = []
    dashers = read_dashers(dasher_fname)
    tasks = read_tasks(task_fname)
    for i in range(10):
        for num_tasks in tasklist_sizes:
            for num_dashers in dasherlist_sizes:
                sim = Baseline()
                system = System(graph)
                chosen_dashers = rand.sample(dashers, num_dashers)
                chosen_tasks = rand.sample(tasks, num_tasks)
                schedule_events(sim, chosen_dashers, chosen_tasks, system, stop_time)
                sim.run()
                num_events = sim.events_processed
                total_reward = system.total_reward
                pct_total_possible = (system.total_reward / system.total_possible_reward) *100
                num_events_per_iter.append(num_events)
                total_reward_per_iter.append(total_reward)
                pct_total_possible_reward_per_iter.append((pct_total_possible))
                
                dist_traveled = []
                for dasher in system.dasher_list:
                    dist_traveled.append(dasher.distance_traveled)
                    
                total_dist_traveled = dist_traveled.sum()
                median_dist_traveled = dist_traveled.median()
                avg_dist_traveled = dist_traveled.mean()
                stdev_dist_traveled = dist_traveled.statistics.stdev()
                
                data.append((num_dashers, num_tasks, num_dashers/num_tasks, num_events, total_reward, pct_total_possible, total_dist_traveled, median_dist_traveled, avg_dist_traveled, stdev_dist_traveled, i))
            i = 0
    with open(results_fname, mode='w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(['num_dashers', 'num_tasks', 'ratio_dashers_tasks', 'num_events', 'total_reward', 'pct_total_possible_reward', 'total_dist', 'median_dist', 'avg_dist', 'stdev_dist', 'iter'])
        writer.writerows(data)
    print(f"Average total reward collected: {sum(total_reward_per_iter) / len(total_reward_per_iter)}")
    print(f"Average percent of total possible reward collected: {sum(pct_total_possible_reward_per_iter) / len(pct_total_possible_reward_per_iter)}")
     
```

```python
def test(dasher_fname, task_fname, stop_time, graph, results_fname, tasklist_sizes, dasherlist_sizes):
    num_events_per_iter = []
    total_reward_per_iter = []
    data = []
    pct_total_possible_reward_per_iter = []
    dashers = read_dashers(dasher_fname)
    tasks = read_tasks(task_fname)
    for num_tasks in tasklist_sizes:
        for num_dashers in dasherlist_sizes:
            sim = Baseline()
            system = System(graph)
            chosen_dashers = rand.sample(dashers, num_dashers)
            chosen_tasks = rand.sample(tasks, num_tasks)
            schedule_events(sim, chosen_dashers, chosen_tasks, system, stop_time)
            sim.run()
            num_events = sim.events_processed
            total_reward = system.total_reward
            pct_total_possible = (system.total_reward / system.total_possible_reward) *100
            num_events_per_iter.append(num_events)
            total_reward_per_iter.append(total_reward)
            pct_total_possible_reward_per_iter.append((pct_total_possible))
            data.append((num_dashers, num_tasks, num_dashers/num_tasks, num_events, total_reward, pct_total_possible))
    i = 0
    with open(results_fname, mode='w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(['num_dashers', 'num_tasks', 'ratio_dashers_tasks', 'num_events', 'total_reward', 'pct_total_possible_reward'])
        writer.writerows(data)
        for num_dashers, num_tasks, ratio, num_events, total_reward, pct_total_possible in data:
            print(f"{i}: Total Reward: {total_reward}, # Events Processed: {num_events}, # Dashers: {num_dashers}, # Tasks: {num_tasks}, ratio dashers:tasks: {ratio}, pct total possible reward: {pct_total_possible}")
            i += 1
    print("----------------------------------")
    print(f"Average total reward collected: {sum(total_reward_per_iter) / len(total_reward_per_iter)}")
    print(f"Average percent of total possible reward collected: {sum(pct_total_possible_reward_per_iter) / len(pct_total_possible_reward_per_iter)}")
     
```
