# Generating file inputs for edge case testing

```python
import csv
import random
```

## Full 100x100 grid for map input

### all with edge weight 1

```python
with open('input/grids/full100-edgeweight1.csv', mode='w', newline='') as file:
        writer = csv.writer(file)
        for i in range(100):
            for j in range(100):
                writer.writerow([i, j, 1])
```

### randomized edge weights

```python
with open('input/grids/full100-edgeweightrand.csv', mode='w', newline='') as file:
        writer = csv.writer(file)
        for i in range(100):
            for j in range(100):
                if i != j:
                    writer.writerow([i, j, random.randint(1, 10)])
```

## Tasks

### All at high end of grid

#### 1 new task per min 750 total time

```python
with open('input/tasks/all-high-1pmin.csv', mode='w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(['VERTEX', 'minute'])
        for i in range(750):
            writer.writerow([random.randint(75, 99), i])
```
