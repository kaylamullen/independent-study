# Analysis of algorithm results

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Diff dasher ratios

```python
sd = pd.read_csv('results10x/smartdispatch/highlow-D1-T5.csv').sort_values(by='ratio_dashers_tasks')
b = pd.read_csv('results10x/baseline/highlow-D1-T5.csv').sort_values(by='ratio_dashers_tasks')
```

## average pct total possible reward by ratio d:t

```python
sd_avgs = sd.groupby('ratio_dashers_tasks')['pct_total_possible_reward'].mean()
b_avgs = b.groupby('ratio_dashers_tasks')['pct_total_possible_reward'].mean()
```

```python
sd_stds = sd.groupby('ratio_dashers_tasks')['pct_total_possible_reward'].std().reset_index()
b_stds = b.groupby('ratio_dashers_tasks')['pct_total_possible_reward'].std().reset_index()
sd_avgs = sd_avgs.reset_index()
b_avgs = b_avgs.reset_index()
```

```python

fig, ax = plt.subplots(figsize=(9, 5))

# Shaded std dev band
ax.fill_between(sd_avgs['ratio_dashers_tasks'], 
                sd_avgs['pct_total_possible_reward'] - sd_stds['pct_total_possible_reward'], 
                sd_avgs['pct_total_possible_reward'] + sd_stds['pct_total_possible_reward'], 
                alpha=0.2, color='steelblue', label='±1 Std Dev')

# Mean line + markers
ax.plot(sd_avgs['ratio_dashers_tasks'], 
        sd_avgs['pct_total_possible_reward'], 
        color='steelblue', linewidth=2, marker='o', markersize=5, label='Mean % Total Reward')


# Shaded std dev band for baseline
ax.fill_between(b_avgs['ratio_dashers_tasks'], 
                b_avgs['pct_total_possible_reward'] - b_stds['pct_total_possible_reward'], 
                b_avgs['pct_total_possible_reward'] + b_stds['pct_total_possible_reward'],
                alpha=0.15, color='tomato', label='±1 Std Dev (Baseline)')

# Baseline mean line + markers
ax.plot(b_avgs['ratio_dashers_tasks'], 
        b_avgs['pct_total_possible_reward'],
        color='tomato', linewidth=2, marker='o',
        markersize=5, linestyle='--', label='Mean % Total Reward (Baseline)')

ax.errorbar(sd_avgs['ratio_dashers_tasks'], sd_avgs['pct_total_possible_reward'],
            yerr=sd_stds['pct_total_possible_reward'], fmt='o', color='steelblue', capsize=4, linewidth=2)


ax.errorbar(b_avgs['ratio_dashers_tasks'], b_avgs['pct_total_possible_reward'],
            yerr=b_stds['pct_total_possible_reward'], fmt='o', color='tomato', capsize=4, linewidth=2)


ax.set_xlabel('Dasher-to-Task Ratio', fontsize=12)
ax.set_ylabel('% Total Possible Reward', fontsize=12)
ax.set_title('Simulation Performance vs. Dasher-to-Task Ratio\n(mean ± 1 std dev, n=10 runs)', fontsize=13)
ax.legend()
ax.grid(True, linestyle='--', alpha=0.5)
ax.set_ylim(0, 105)

plt.tight_layout()
plt.show()

```
