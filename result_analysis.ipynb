# Analysis of algorithm results

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Diff dasher ratios

```python
sd = pd.read_csv('results/smartdispatch/D1-T200.csv').sort_values(by='ratio_dashers_tasks')
b = pd.read_csv('results/baseline/D1-T200.csv').sort_values(by='ratio_dashers_tasks')
```

```python
sd
```

```python
b.head()
```

```python
# plot lines
plt.plot(sd['ratio_dashers_tasks'], sd['pct_total_possible_reward'], label = "smart dispatch")
plt.plot(b["ratio_dashers_tasks"], b['pct_total_possible_reward'], label = "baseline")
plt.xlabel("ratio of dashers to tasks")
plt.ylabel("percent of total possible reward")
plt.legend()
plt.show()
```

```python
# plot lines
plt.plot(sd['ratio_dashers_tasks'], sd['pct_total_possible_reward'], label = "smart dispatch")
plt.plot(b["ratio_dashers_tasks"], b['pct_total_possible_reward'], label = "baseline")
plt.xlim(0, 0.2)
plt.xlabel("ratio of dashers to tasks")
plt.ylabel("percent of total possible reward")
plt.legend()
plt.show()
```

```python
# plot lines
plt.plot(sd['ratio_dashers_tasks'], sd['pct_total_possible_reward'], label = "smart dispatch")
plt.plot(b["ratio_dashers_tasks"], b['pct_total_possible_reward'], label = "baseline")
plt.xlim(-0.001, 0.1)
plt.xlabel("ratio of dashers to tasks")
plt.ylabel("percent of total possible reward")
plt.legend()
plt.show()
```

```python
# plot lines
plt.plot(sd['ratio_dashers_tasks'], sd['pct_total_possible_reward'], label = "smart dispatch")
plt.plot(b["ratio_dashers_tasks"], b['pct_total_possible_reward'], label = "baseline")
plt.xlim(0, 0.01)
plt.xlabel("ratio of dashers to tasks")
plt.ylabel("percent of total possible reward")
plt.legend()
plt.show()
```
